```python
# === 0) MOUNT DRIVE & UNZIP DATASET ===
from google.colab import drive
drive.mount('/content/drive', force_remount=True)

import zipfile, os
zip_path     = '/content/drive/MyDrive/portable-ai-based-sign-language-translator-export.zip'
extract_path = '/content/drive/MyDrive/portable_sign_language'
os.makedirs(extract_path, exist_ok=True)
with zipfile.ZipFile(zip_path, 'r') as z:
    z.extractall(extract_path)
print("Extraction complete!")

# === 1) IMPORTS ===
import json, numpy as np, pandas as pd, cv2, matplotlib.pyplot as plt, tensorflow as tf
from sklearn.model_selection import train_test_split

# === 2) LOAD LABELS JSON & BUILD DATAFRAME ===
labels_path = os.path.join(extract_path, 'info.labels')
with open(labels_path, 'r') as f:
    info = json.load(f)

df = pd.DataFrame(info['files'])
# Extract just the gesture name (e.g. "00015_agree" → "agree")
df['label_clean'] = df['name'].apply(lambda x: x.split('_',1)[1])
df['full_path']  = df['path'].apply(lambda p: os.path.join(extract_path, p))

# === 3) FILTER OUT MISSING/UNREADABLE IMAGES ===
def is_valid(p):
    if not os.path.exists(p): return False
    img = cv2.imread(p)
    return img is not None

mask = df['full_path'].apply(is_valid)
print(f"Discarding {len(df)-mask.sum()} unreadable images")
df = df[mask].reset_index(drop=True)

# === 4) SPLIT INTO TRAIN / TEST ===
train_df, test_df = train_test_split(
    df, test_size=0.2, random_state=42, shuffle=True
)
print(f"Train samples: {len(train_df)}   Test samples: {len(test_df)}")
print("Train class dist:\n", train_df['label_clean'].value_counts())
print("Test  class dist:\n", test_df ['label_clean'].value_counts())

# === 5) PREPROCESSING FUNCTION ===
IMG_SIZE = 128

def preprocess_image(path, augment=False):
    img = cv2.imread(path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
    img = img.astype(np.float32) / 255.0
    if augment:
        # horizontal flip
        if np.random.rand() < 0.5:
            img = np.fliplr(img)
        # small rotation
        ang = np.random.uniform(-15,15)
        M   = cv2.getRotationMatrix2D((IMG_SIZE/2,IMG_SIZE/2), ang, 1.0)
        img = cv2.warpAffine(img, M, (IMG_SIZE, IMG_SIZE))
    return img

# === 6) PREPARE GENERATORS ===
classes      = sorted(df['label_clean'].unique())
lbl2idx      = {c:i for i,c in enumerate(classes)}
num_classes  = len(classes)

train_paths  = train_df['full_path'].tolist()
train_labels = train_df['label_clean'].map(lbl2idx).tolist()
test_paths   = test_df ['full_path'].tolist()
test_labels  = test_df ['label_clean'].map(lbl2idx).tolist()

def data_generator(paths, labels, batch_size=32, augment=False):
    n = len(paths)
    while True:
        idxs = np.random.permutation(n)
        for i in range(0, n, batch_size):
            batch = idxs[i:i+batch_size]
            X, y = [], []
            for j in batch:
                img = preprocess_image(paths[j], augment=augment)
                X.append(img)
                onehot = np.zeros(num_classes, dtype=np.float32)
                onehot[labels[j]] = 1.0
                y.append(onehot)
            yield np.stack(X), np.stack(y)

batch_size      = 32
train_gen       = data_generator(train_paths,  train_labels, batch_size, augment=True)
test_gen        = data_generator(test_paths,   test_labels,  batch_size, augment=False)
steps_per_epoch = len(train_paths)//batch_size
validation_steps= len(test_paths)//batch_size

# === 7) BUILD & COMPILE CNN ===
model = tf.keras.Sequential([
    tf.keras.layers.Input((IMG_SIZE,IMG_SIZE,3)),
    tf.keras.layers.Conv2D(32,3,activation='relu',padding='same'),
    tf.keras.layers.MaxPooling2D(),
    tf.keras.layers.Conv2D(64,3,activation='relu',padding='same'),
    tf.keras.layers.MaxPooling2D(),
    tf.keras.layers.Conv2D(128,3,activation='relu',padding='same'),
    tf.keras.layers.MaxPooling2D(),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128,activation='relu'),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(num_classes,activation='softmax'),
])
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
model.summary()

# === 8) TRAIN & PLOT METRICS ===
epochs = 10
history = model.fit(
    train_gen,
    steps_per_epoch=steps_per_epoch,
    validation_data=test_gen,
    validation_steps=validation_steps,
    epochs=epochs
)

# Plot training & validation curves
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(history.history['loss'], label='train_loss')
plt.plot(history.history['val_loss'], label='val_loss')
plt.legend(); plt.title('Loss')
plt.subplot(1,2,2)
plt.plot(history.history['accuracy'], label='train_acc')
plt.plot(history.history['val_accuracy'], label='val_acc')
plt.legend(); plt.title('Accuracy')
plt.show()

# === 9) EVALUATE on TEST SET ===
loss, acc = model.evaluate(test_gen, steps=validation_steps)
print(f"Test Accuracy: {acc*100:.2f}%")

# === 10) SAVE KERAS MODEL ===
out_dir = '/content/drive/MyDrive/models'
os.makedirs(out_dir, exist_ok=True)
h5_path = os.path.join(out_dir, 'sign_language_cnn.h5')
model.save(h5_path)
print(f"Keras model saved to {h5_path}")

# === 11) EXPORT FP32 TFLITE ===
fp32_tflite = tf.lite.TFLiteConverter.from_keras_model(model).convert()
with open('model_fp32.tflite','wb') as f:
    f.write(fp32_tflite)
print("FP32 TFLite model saved to model_fp32.tflite")

# === 12) EXPORT INT8 QUANTIZED TFLITE ===
def rep_data_gen():
    for p in train_paths[:100]:
        img = preprocess_image(p).astype(np.float32)
        yield [np.expand_dims(img,0)]

int8_conv = tf.lite.TFLiteConverter.from_keras_model(model)
int8_conv.optimizations = [tf.lite.Optimize.DEFAULT]
int8_conv.representative_dataset = rep_data_gen
int8_conv.target_spec.supported_ops = [tf.lite.OpsSet.TFLITE_BUILTINS_INT8]
int8_conv.inference_input_type  = tf.int8
int8_conv.inference_output_type = tf.int8
tflite_int8 = int8_conv.convert()
with open('model_int8.tflite','wb') as f:
    f.write(tflite_int8)
print("INT8 TFLite model saved to model_int8.tflite")

# === 13) COMPARE FILE SIZES ===
fp32_kb = os.path.getsize('model_fp32.tflite')/1024
int8_kb = os.path.getsize('model_int8.tflite')/1024
print(f"FP32 TFLite size: {fp32_kb:.1f} KB")
print(f" INT8 TFLite size: {int8_kb:.1f} KB")

# === 14) EVALUATE INT8 MODEL ACCURACY ===
interpreter = tf.lite.Interpreter(model_path='model_int8.tflite')
interpreter.allocate_tensors()
inp_det   = interpreter.get_input_details()[0]
out_det   = interpreter.get_output_details()[0]
in_scale, in_zero = inp_det['quantization']
out_scale, out_zero = out_det['quantization']

y_true, y_pred_int8 = [], []
for p, lbl in zip(test_paths, test_labels):
    img = preprocess_image(p).astype(np.float32)
    qin = (img/in_scale + in_zero).round().astype(np.int8)
    interpreter.set_tensor(inp_det['index'], np.expand_dims(qin,0))
    interpreter.invoke()
    qout = interpreter.get_tensor(out_det['index'])[0].astype(np.int32)
    logits = (qout - out_zero) * out_scale
    y_true.append(lbl)
    y_pred_int8.append(np.argmax(logits))

y_true = np.array(y_true)
y_pred_int8 = np.array(y_pred_int8)
acc_int8 = (y_true == y_pred_int8).mean()
print(f"INT8 Quantized model accuracy: {acc_int8*100:.2f}%")
```

```python
# =========================================================
# 0)  MOUNT GOOGLE DRIVE & EXTRACT ZIP
# =========================================================
from google.colab import drive; drive.mount('/content/drive', force_remount=True)
import os, zipfile, json, tensorflow as tf, numpy as np, pandas as pd

ZIP  = '/content/drive/MyDrive/portable-ai-based-sign-language-translator-export.zip'
ROOT = '/content/drive/MyDrive/portable_sign_language'
os.makedirs(ROOT, exist_ok=True)
with zipfile.ZipFile(ZIP) as z: z.extractall(ROOT)
print("Data extracted ➜", ROOT)

# =========================================================
# 1)  READ LABELS JSON  →  DATAFRAME
# =========================================================
meta = json.load(open(os.path.join(ROOT, 'info.labels')))
df   = pd.DataFrame(meta['files'])
df['label_clean'] = df['name'].apply(lambda s: s.split('_',1)[1])
df['full_path']   = df['path'].apply(lambda p: os.path.join(ROOT, p))

# keep only readable images
df = df[df['full_path'].apply(lambda p: tf.io.gfile.exists(p))].reset_index(drop=True)
classes = sorted(df['label_clean'].unique())
lbl2idx = {c:i for i,c in enumerate(classes)}
print(f"Images: {len(df)}  •  Classes: {len(classes)}")

# =========================================================
# 2)  STRATIFIED TRAIN / TEST SPLIT
# =========================================================
from sklearn.model_selection import train_test_split
train_df, test_df = train_test_split(
    df, test_size=0.2, stratify=df['label_clean'], random_state=42)

train_paths, y_train = train_df['full_path'].tolist(), train_df['label_clean'].map(lbl2idx).tolist()
test_paths , y_test  = test_df ['full_path'].tolist(), test_df ['label_clean'].map(lbl2idx).tolist()

# =========================================================
# 3)  tf.data INPUT PIPELINE
# =========================================================
IMG  = 128
BATCH= 32
AUTOTUNE = tf.data.AUTOTUNE
NCLASS   = len(classes)

def preprocess(path, label, augment=False):
    img = tf.io.read_file(path)
    img = tf.image.decode_jpeg(img, channels=3)
    img = tf.image.resize(img, [IMG, IMG])
    if augment:
        img = tf.image.random_flip_left_right(img)
    img = tf.cast(img, tf.float32) / 255.0
    return img, tf.one_hot(label, NCLASS)

def make_ds(paths, labels, shuffle=False, augment=False):
    ds = tf.data.Dataset.from_tensor_slices((paths, labels))
    if shuffle: ds = ds.shuffle(len(paths))
    ds = ds.map(lambda p,l: preprocess(p,l,augment), num_parallel_calls=AUTOTUNE)
    return ds.batch(BATCH).prefetch(AUTOTUNE)

train_ds = make_ds(train_paths, y_train, shuffle=True , augment=True)
test_ds  = make_ds(test_paths , y_test , shuffle=False, augment=False)

# =========================================================
# 4)  BUILD MOBILENETV2 α=0.5
# =========================================================
# --- build MobileNetV2 with narrower width ----------------------
base = tf.keras.applications.MobileNetV2(
        input_shape=(IMG,IMG,3),
        alpha=0.35,          # ← 0.35 instead of 0.50
        include_top=False,
        weights='imagenet',
        pooling='avg')
base.trainable = False


model = tf.keras.Sequential([
    base,
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(NCLASS, activation='softmax')
])

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# ---------------- STAGE-1 : head only ----------------
model.fit(train_ds, epochs=5, validation_data=test_ds)

# ---------------- STAGE-2 : fine-tune last 30 layers --
base.trainable = True
for layer in base.layers[:-30]:
    layer.trainable = False

model.compile(optimizer=tf.keras.optimizers.Adam(1e-4),
              loss='categorical_crossentropy',
              metrics=['accuracy'])
model.fit(train_ds, epochs=5, validation_data=test_ds)

# =========================================================
# 5)  FINAL ACCURACY
# =========================================================
loss_fp32, acc_fp32 = model.evaluate(test_ds, verbose=0)
print(f"\nFP32 test accuracy: {acc_fp32*100:.2f}%")

# =========================================================
# 6)  SAVE FP32 & INT8 TFLite
# =========================================================
# FP32
fp32 = tf.lite.TFLiteConverter.from_keras_model(model).convert()
open('model_fp32.tflite','wb').write(fp32)

# ---------- build INT-8 model that matches 96×96  ----------
def representative():
    for p in train_paths[:200]:
        img,_ = preprocess(p, 0, augment=False)   # 96×96 float32 0-1
        yield [img[None]]

conv = tf.lite.TFLiteConverter.from_keras_model(model)
conv.optimizations = [tf.lite.Optimize.DEFAULT]
conv.representative_dataset = representative
conv.target_spec.supported_ops = [tf.lite.OpsSet.TFLITE_BUILTINS_INT8]
conv.inference_input_type = conv.inference_output_type = tf.int8
int8_tflite = conv.convert()

with open('model_int8.tflite', 'wb') as f:
    f.write(int8_tflite)
print("Saved 96×96 INT-8 model:", len(int8_tflite)/1024, "KB")


```

```python
# ======================================================================
# CONFUSION MATRIX  +  TRAINING / VALIDATION CURVES
# ======================================================================
import numpy as np, matplotlib.pyplot as plt, seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report

# ---------- A)  collect FP32 predictions --------------------------------
y_true, y_pred = [], []
for batch_imgs, batch_lbls in test_ds:
    logits = model.predict(batch_imgs, verbose=0)
    y_true.extend(np.argmax(batch_lbls.numpy(), axis=1))
    y_pred.extend(np.argmax(logits,               axis=1))

y_true = np.array(y_true);  y_pred = np.array(y_pred)

print("===  FP32  CLASSIFICATION REPORT  ===\n")
print(classification_report(y_true, y_pred, target_names=classes))

# ---------- B)  plot confusion matrix -----------------------------------
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=classes, yticklabels=classes)
plt.xlabel("Predicted"); plt.ylabel("True"); plt.title("FP32 Confusion Matrix")
plt.tight_layout()
plt.show()

# ---------- C)  (optional) INT-8 predictions ----------------------------
want_int8 = True    # set False if you don’t need this part
if want_int8:
    interp = tf.lite.Interpreter('model_int8.tflite')
    interp.allocate_tensors()
    inp  = interp.get_input_details()[0]
    out  = interp.get_output_details()[0]
    s_i, z_i = inp['quantization']
    s_o, z_o = out['quantization']

    y_pred_int8 = []
    for p in test_paths:
        # load & preprocess (float32 0-1)
        img = tf.io.read_file(p)
        img = tf.image.decode_jpeg(img, channels=3)
        img = tf.image.resize(img, [IMG, IMG])
        img = tf.cast(img, tf.float32) / 255.0
        # quantise → int8 and infer
        q = tf.cast(tf.round(img / s_i + z_i), tf.int8)[None]
        interp.set_tensor(inp['index'], q.numpy())
        interp.invoke()
        qout = interp.get_tensor(out['index'])[0]
        prob = (qout.astype(np.int32) - z_o) * s_o
        y_pred_int8.append(np.argmax(prob))

    print("\n===  INT-8  CLASSIFICATION REPORT  ===\n")
    print(classification_report(y_true, y_pred_int8, target_names=classes))

    # confusion matrix for INT-8
    cm8 = confusion_matrix(y_true, y_pred_int8)
    plt.figure(figsize=(8,6))
    sns.heatmap(cm8, annot=True, fmt='d', cmap='Greens',
                xticklabels=classes, yticklabels=classes)
    plt.xlabel("Predicted"); plt.ylabel("True"); plt.title("INT-8 Confusion Matrix")
    plt.tight_layout()
    plt.show()


# ---------- E)  print INT-8 file size -----------------------------------
int8_kb = os.path.getsize('model_int8.tflite') / 1024
print(f"\nINT-8 model file size : {int8_kb:.1f} KB")
```
